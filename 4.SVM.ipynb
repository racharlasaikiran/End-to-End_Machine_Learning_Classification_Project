```python
import numpy as np
import pandas as pd
import seaborn as sb
import matplotlib.pyplot as plt
import warnings
warnings.simplefilter('ignore')
```

```python
pd.set_option('display.max_columns',None)
```

```python
x_train = pd.read_csv('x_train_data.csv')
x_test = pd.read_csv('x_test_data.csv')
```

```python
y_train = pd.read_csv('y_train_data.csv')
y_test = pd.read_csv('y_test_data.csv')
```

```python
from sklearn.decomposition import PCA
pc = PCA(n_components=0.95)
x_train=pc.fit_transform(x_train)
x_test = pc.transform(x_test)
```

```python
pc.explained_variance_ratio_ *100
```

```python
from sklearn.svm import SVC
from sklearn.model_selection import GridSearchCV
estimator = SVC()
param_grid = {"C":[0.01,0.1,1,10,100],'kernel':['linear','rbf','poly','sigmoid']}
gscv = GridSearchCV(estimator,param_grid,cv=5,scoring='accuracy')
gscv.fit(x_train,y_train)
print("Best k:", gscv.best_params_)
```

```python
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score

#modeling
model = SVC(C=0.01,kernel='linear')
model.fit(x_train,y_train)

#predection
x_train_pred = model.predict(x_train)
x_test_pred = model.predict(x_test)

#evolution
print('train accuracy :', accuracy_score(y_train, x_train_pred))
print('test accuracy :', accuracy_score(y_test, x_test_pred))
from sklearn.model_selection import cross_val_score
print('cv score :',cross_val_score(model,x_train,y_train,cv=5,scoring='accuracy').mean())
```
