```python
import numpy as np
import pandas as pd
import tensorflow
import keras
import warnings
warnings.simplefilter('ignore')
```

```python
pd.set_option('display.max_columns',None)
```

```python
x_train = pd.read_csv('x_train_data.csv')
x_test = pd.read_csv('x_test_data.csv')
```

```python
y_train = pd.read_csv('y_train_data.csv')
y_test = pd.read_csv('y_test_data.csv')
```

```python
from sklearn.decomposition import PCA
pc = PCA(n_components=0.95)
x_train=pc.fit_transform(x_train)
x_test = pc.transform(x_test)
```

```python
pc.explained_variance_ratio_ *100
```

```python
from keras.models import Sequential
from keras.layers import Dense
ann = Sequential()
ann.add(Dense(input_dim =15 , units =29 ,kernel_initializer ='uniform', activation= 'relu'))
ann.add(Dense(units =29 ,kernel_initializer ='uniform', activation= 'relu'))
ann.add(Dense(units =29 ,kernel_initializer ='uniform', activation= 'relu'))
ann.add(Dense(units =29 ,kernel_initializer ='uniform', activation= 'relu'))
ann.add(Dense(units =29 ,kernel_initializer ='uniform', activation= 'relu'))
ann.add(Dense(units =1 ,kernel_initializer ='uniform', activation= 'sigmoid'))
ann.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])
```

```python
ann.fit(x_train,y_train,batch_size=20,epochs=500)
```

```python
# Predict probabilities
x_train_pred = ann.predict(x_train)
x_test_pred  = ann.predict(x_test)

x_train_pred = (x_train_pred >= 0.5).astype(int)
x_test_pred  = (x_test_pred  >= 0.5).astype(int)
```

```python
from sklearn.metrics import accuracy_score
print("Training Accuracy: ",accuracy_score(y_train, x_train_pred))
print("Testing Accuracy: ",accuracy_score(y_test, x_test_pred))
```
